<h2>1. Descripcion General</h2>

<h3>1.1.  Filosofia</h3> 

Markdown pretende ser tan fácil de leer y escribir como sea posible. 

La legibilidad, sin embargo, se enfatiza por encima de todo lo demás. Un 
documento con formato Markdown debe poder publicarse tal como está, como texto sin formato, sin que 
parezca que se ha marcado con etiquetas o instrucciones de formato. Si bien 
la sintaxis de Markdown ha sido influenciada por varios 
filtros de texto a HTML existentes , incluyendo [Setext] [1], [atx] [2], [Textile] [3], [reStructuredText] [4], 
[Grutatext] [ 5], y [EtText] [6]: la mayor fuente de 
inspiración para la sintaxis de Markdown es el formato de texto plano del correo electrónico . 

  [1]: http://docutils.sourceforge.html
  [2]: http://www.aaronsw.com/2002/atx/ 
  [3]: http://textism.com/tools/textile/ 
  [4]: http://docutils.sourceforge.net/rst.html 
  [5]: http://www.triptico.com/software/grutatxt.html 
  [6]: http://ettext.taint.org/doc/ 

Para este fin, la sintaxis de Markdown se compone completamente de 
caracteres de puntuación , que los caracteres de puntuación han sido elegidos cuidadosamente 
para parecerse a lo que significan. Por ejemplo, los asteriscos alrededor de una palabra en realidad se 
ven como \ * énfasis \ *. Las listas de rebajas parecen, bueno, listas. Incluso las 
citas en bloque parecen pasajes de texto citados, suponiendo que alguna vez hayas 
usado el correo electrónico. 

<h3>1.2. HTML en línea </h3> 

La sintaxis de Markdown está diseñada para un propósito: para ser usada como
Formato para * escritura * para la web. 

Markdown no es un reemplazo de HTML, ni siquiera está cerca de él. Su 
sintaxis es muy pequeña, correspondiendo solo a un subconjunto muy pequeño de 
etiquetas HTML. La idea es * no * crear una sintaxis que facilite la 
inserción de etiquetas HTML. En mi opinión, las etiquetas HTML ya son fáciles de 
insertar. La idea de Markdown es facilitar la lectura, escritura y 
edición de la prosa. HTML es un formato de * publicación *; Markdown es un 
formato de * escritura * . Por lo tanto, la sintaxis de formato de Markdown solo aborda los problemas que 
se pueden transmitir en texto sin formato. 

Para cualquier marca que no esté cubierta por la sintaxis de Markdown, simplemente 
use el propio HTML. No hay necesidad de prefaciarlo o delimitarlo.
indica que estás cambiando de Markdown a HTML; solo usas 
las etiquetas 

Las únicas restricciones son que los elementos HTML a nivel de bloque (por ejemplo, `<div>`, 
`<table>`, `<pre>`, `<p>`, etc.) deben estar separados del 
contenido circundante con líneas en blanco , y las etiquetas de inicio y fin del bloque 
no deben estar sangradas con tabulaciones o espacios. Markdown es lo suficientemente inteligente como 
para no agregar etiquetas `<p>` adicionales (no deseadas) alrededor de las etiquetas HTML a nivel de bloque. 

Por ejemplo, para agregar una tabla HTML a un artículo de Markdown: 

    Este es un párrafo regular. 

    <table> 
        <tr> 
            <td> Foo </td> 
        </tr> 
    </table>


Tenga en cuenta que la sintaxis de formato Markdown no se procesa dentro de las 
etiquetas HTML a nivel de bloque . Por ejemplo, no puede usar el estilo * * de Markdown dentro de un 
bloque HTML. 

Las etiquetas HTML de nivel de extensión (por ejemplo, `<span>`, `<cite>`, o `<del>`) se pueden 
usar en cualquier lugar en un párrafo, un elemento de lista o un encabezado de Markdown. Si lo 
desea, puede incluso usar etiquetas HTML en lugar del formato Markdown; por ejemplo, si 
prefiere usar etiquetas HTML `<a>` o `<img>` en lugar de la 
sintaxis de enlace o imagen de Markdown , siga adelante. 

A diferencia de las etiquetas HTML de nivel de bloque, la sintaxis Markdown * se * procesa dentro 
de las etiquetas de nivel de rango. 



<h3>1.3. Escape automático para caracteres especiales </h3> 

En HTML,
y `&`. Los paréntesis angulares izquierdos se utilizan para iniciar etiquetas; ampersands se 
utilizan para denotar las entidades HTML. Si desea usarlos como 
caracteres literales , debe evitarlos como entidades, por ejemplo, `& lt;`, y 
`& amp;`. 

Los símbolos comerciales en particular son un problema para los escritores web. Si desea 
escribir sobre 'AT&T', debe escribir '`AT & amp; T`'. Incluso necesitas 
escapar de ampersands dentro de las URLs. Por lo tanto, si desea vincular a: 

    http://images.google.com/images?num=30&q=larry+bird 

necesita codificar la URL como: 

    http://images.google.com/images?num= 30 & amp; q = larry + bird 

en el atributo `href` de la etiqueta de anclaje. No hace falta decir que esto es fácil de 
olvidar.
Errores en sitios web bien marcados. 

Markdown te permite usar estos personajes de forma natural, cuidando de 
todos los escapes necesarios para ti. Si usa un signo y como parte de 
una entidad HTML, no se modifica; De lo contrario, se traducirá 
a `& amp;`. 

Por lo tanto, si desea incluir un símbolo de copyright en su artículo, puede escribir: 

    & copiar; 

y Markdown lo dejará solo. Pero si escribe: 

    AT&T 

Markdown lo traducirá a: 

    AT & amp; T 

Del mismo modo, porque Markdown es compatible con [HTML en línea] (# html), si utiliza 
corchetes angulares como delimitadores para etiquetas HTML, Markdown los tratará como 
tales. Pero si escribes: 

    4 <5

Markdown lo traducirá a: 

    4 & lt; 5 

Sin embargo, los intervalos y bloques dentro del código Markdown, los corchetes angulares y los 
símbolos & * siempre se codifican automáticamente. Esto facilita el uso de 
Markdown para escribir sobre el código HTML. (A diferencia del código HTML sin procesar, que es un 
formato terrible para escribir sobre la sintaxis HTML, porque cada uno de los 
códigos "<` y `&` en su código de ejemplo debe escaparse.)

<H2>2. Elementos de Bloque </h2> 

<h3>2.1. Párrafos y saltos de línea </h3> 

Un párrafo es simplemente una o más líneas de texto consecutivas, separadas 
por una o más líneas en blanco.
en blanco.) Los párrafos normales no deben estar sangrados con espacios o tabulaciones. 

La implicación de la regla de "una o más líneas consecutivas de texto" es 
que Markdown admite párrafos de texto "rígidos". Esto difiere 
significativamente de la mayoría de los demás formateadores de texto a HTML (incluida 
la opción "Convertir saltos de línea" de Movable Type) que traduce cada 
carácter de salto de línea en un párrafo en una etiqueta `<br />`. 

Cuando * quieres * insertar una etiqueta break de `<br />` con Markdown, 
finalizas una línea con dos o más espacios y luego escribes return. 

Sí, esto requiere un poco más de esfuerzo para crear un `<br />`, pero un simple 
"cada salto de línea es una regla` <br /> `" no funcionaría para Markdown.
El estilo de correo electrónico de Markdown [blockquoting] [bq] y los [párrafos de la lista] [l] de varios párrafos 
funcionan mejor, y se ven mejor, cuando los formatea con cortes duros. 

  [bq]: #blockquote 
  [l]: #list 

<h3>2.2. Encabezado o Headers </h3> 

Markdown admite dos estilos de encabezados, [Setext] [1] y [atx] [2]. 

Los encabezados de estilo setext están "subrayados" usando signos de igual (para 
encabezados de primer nivel ) y guiones (para encabezados de segundo nivel). Por ejemplo: 

    Este es un H1 
    ============= 

    Este es un H2 
    ------------- 

Cualquier número de subrayado `=` 's o `-` va a funcionar. 

Los encabezados de estilo Atx usan 1-6 caracteres hash al comienzo de la línea, 
correspondientes a los niveles de encabezado 1-6.

    Código Markdown:
~~~    
    # Este es un H1 
    ## Este es un H2
    ### Este es un H3 
    #### Este es un H4
    ##### Este es un H5
    ###### Este es un H6 
~~~
    Salida:
    
# Este es un H1 
## Este es un H2
### Este es un H3 
#### Este es un H4
##### Este es un H5
###### Este es un H6     
    
Opcionalmente, puede "cerrar" los encabezados de estilo atx. Esto es puramente 
cosmético, puedes usarlo si crees que se ve mejor. Los 
hashes de cierre ni siquiera tienen que coincidir con la cantidad de hashes 
utilizados para abrir el encabezado. (El número de hashes de apertura 
determina el nivel del encabezado.): 

    # Este es un H1 # 

    ## Este es un H2 ## 

    ### Este es un H3 ###### 

<h3>2.3. Comentar párrafos, citas o Blockquotes </h3> 

Markdown usa caracteres de estilo de correo electrónico `>` para la asignación de bloques. Si está 
familiarizado con citar pasajes de texto en un mensaje de correo electrónico, entonces 
sabe cómo crear una cita en bloque en Markdown.
ajuste el texto y ponga un `>` antes de cada línea: 

    código markdown:
~~~    
    > Esta es una cita con dos párrafos. Lorem ipsum dolor sit amet, 
    consectetuer adipiscing elit. Aliquam hendrerit mi posuere lectus. 
    > Vestibulum enim wisi, viverra nec, fringilla in, laoreet vitae, risus. 
    >> 
    Donec sit amet nisl. Aliquam sempre ipsum sit amet velit. Suspendido 
    > id sem consectetuer libero luctus adipiscing. 
~~~    
    Salida:

> Esta es una cita con dos párrafos. Lorem ipsum dolor sit amet,  
consectetuer adipiscing elit. Aliquam hendrerit mi posuere lectus.   
> Vestibulum enim wisi, viverra nec, fringilla in, laoreet vitae, risus.  
>>  
Donec sit amet nisl. Aliquam sempre ipsum sit amet velit. Suspendido  
> id sem consectetuer libero luctus adipiscing.  

Markdown le permite ser perezoso y solo colocar `>` antes de la primera 
línea de un párrafo completo: 

    código markdown:
~~~

    consultor adipiscing elit. Aliquam hendrerit mi posuere lectus.

Las citas de bloque pueden anidarse (es decir, una cita de blockquote-in-a-block)

    código markdown:
~~~

<h3>2.4. Listas </h3> 

Markdown admite listas ordenadas (numeradas) y no ordenadas (con viñetas). 

Las __`listas desordenadas`__ usan asteriscos, ventajas y guiones, intercambiables

    Código Markdown:
~~~

    * Verde 
    * Azul 
~~~
    Salida:
    
* Rojo 
* Verde 
* Azul     
_____    
es equivalente a: 
_____
    Código Markdown:

`+ Rojo`  
`+ Verde`  
`+ Azul`  

    Salida:
    
+ Rojo 
+ Verde 
+ Azul     


Las __`listas ordenadas`__ usan números seguidos por un punto y un espacio:  

`1. Pájaro`   
`2. McHale`   
`3. Parroquia` 
    
    Salida:
    
1. Pájaro 
2. McHale 
3. Parroquia     
    
Es importante tener en cuenta que los números reales que usa para marcar la 
lista no tienen efecto en la salida HTML que produce Markdown. 

Si en cambio escribiste la lista en Markdown como esto: 

1. Bird 
1. McHale 
1. Parish 

o incluso: 

3. Bird 
1. McHale 
8. Parish 

obtendrías exactamente la misma salida HTML. El punto es, si quieres,
puede usar números ordinales en sus listas de Markdown ordenadas, de modo que 
los números en su fuente coincidan con los números en su HTML publicado. 
Pero si quieres ser perezoso, no tienes que hacerlo. 

Sin embargo, si usa una numeración de lista lenta, aún debe comenzar la 
lista con el número 1. En algún momento en el futuro, Markdown puede admitir el 
inicio de listas ordenadas en un número arbitrario.

<h3>2.5. Bloques de código </h3> 

Los bloques de código preformateados se utilizan para escribir sobre programación o 
código fuente de marcado. En lugar de formar párrafos normales, las líneas 
de un bloque de código se interpretan literalmente. Markdown envuelve un bloque de código 
en las etiquetas `<pre>` y `<code>`. 

Para producir un bloque de código en Markdown, simplemente sangra cada línea del 
bloque con al menos 4 espacios o 1 pestaña. Por ejemplo, dada esta entrada: 

    Este es un párrafo normal: 

        este es un bloque de código. 

con tres `````
~~~
```lenguaje
Codigo que quiero
en varias lineas
```
~~~
salida:
Ejemplo python

```python
import lifetime
for each_day in lifetime.days():
    carpe diem()
```

Ejemplo R
```r
group <- gl( 2, 10, 20,
labels = c( "Ctl", "Trt" ) )
weight <- c( ctl, trt )
lm.D9 <- lm( weight ~ group )
lm.D90 <- lm( weight ~ group
- 1 ) # omitting intercept
```
o tambien con tres `~~~`
```
~~~
Codigo que quiero
en varias lineas
~~~
```
salida:

~~~
Codigo que quiero
en varias lineas
~~~

o tambien con tres ---

Dentro de un bloque de código, los símbolos & ('& \`) y los corchetes angulares (` <`y`> `) 
se convierten automáticamente en entidades HTML. Esto lo hace muy
fácil de incluir código fuente HTML de ejemplo usando Markdown: solo pégalo 
y sangra, y Markdown se encargará de la molestia de codificar los 
símbolos y los paréntesis angulares. Por ejemplo, esto: 

        <div class = "footer"> 
            & copy; 2004 Foo Corporation 
        </div> 

se convertirá en: 

    <pre> <code> & lt; div class = "footer" & gt; 
        & amp; copia; 2004 Foo Corporation 
    & lt; / div & gt; 
    </code> </pre> 

La sintaxis de Markdown normal no se procesa dentro de los bloques de código. Por ejemplo, los 
asteriscos son solo asteriscos literales dentro de un bloque de código. Esto significa

<h3>3.1. Links </h3> 

Markdown admite dos estilos de enlaces : `inline` y `reference`. 

En ambos estilos, el texto del enlace está delimitado por [corchetes]. 

1. Para crear un __enlace en línea__, use un conjunto de paréntesis regulares inmediatamente tras el corchete de cierre del texto del enlace. Dentro de los paréntesis, coloque la URL donde desea que apunte el enlace, junto con un título * opcional * para el enlace, entre comillas. Por ejemplo: 

    Esto es [un ejemplo] (http://example.com/ "Título") enlace en línea. 

    [Este enlace] (http://example.net/) no tiene ningún atributo de título. 

    Producirá: 

Esto es [un ejemplo] (http://example.com/ "Título") enlace en línea. 

[Este enlace] (http://example.net/) no tiene ningún atributo de título.     

2. Si se refiere a un `recurso local` en el mismo servidor, puede usar __rutas relativas__:

Los *enlaces de estilo de referencia* utilizan un segundo conjunto de corchetes, dentro del 
cual coloca una etiqueta de su elección para identificar el enlace: 

    Este es un enlace de estilo de referencia [ejemplo] [id]. 

Opcionalmente, puede usar un espacio para separar los conjuntos de corchetes: 

    Esto es [un ejemplo] [id] enlace de estilo de referencia. 

Luego, en cualquier parte del documento, define su etiqueta de enlace como esta, 
en una línea por sí misma: 

    [id]: http://example.com/ "Título opcional aquí" 

salida:

Este es un enlace de estilo de referencia [ejemplo] [id].  
Esto es [un ejemplo] [id] enlace de estilo de referencia.  

[id]: http://example.com/ "Título opcional aquí" 

Es decir: 

* Corchetes que contienen el identificador del enlace (opcionalmente 
    sangrado desde el margen izquierdo usando hasta tres espacios); 
* seguido de dos puntos; 
* seguido de uno o más espacios (o pestañas);
* seguido de la URL del enlace; 
* Opcionalmente, seguido de un atributo de título para el enlace, 
    entre comillas dobles o simples, o entre paréntesis. 

3. Las siguientes tres definiciones de enlaces son equivalentes: 

	[foo]: http://example.com/ "Título opcional aquí" 
	[foo]: http://example.com/ 'Título opcional aquí' 
	[foo]: http: // example.com/ (Título opcional aquí) 

** Nota: ** Hay un error conocido en Markdown.pl 1.0.1 que evita que 
se usen comillas simples para delimitar títulos de enlaces. 

La URL del enlace puede, opcionalmente, estar rodeada por corchetes angulares: 

    [id]: <http://example.com/> "Título opcional aquí" 

Puede poner el atributo del título en la siguiente línea y usar espacios adicionales
o pestañas para el relleno, que tienden a verse mejor con las URL más largas: 

    [id]: http://example.com/longish/path/to/resource/here 
        "Título opcional aquí" Las 

definiciones de los enlaces solo se usan para crear enlaces durante Markdown 
Procesando, y se eliminan de su documento en la salida HTML. 

Los nombres de definición de enlaces pueden estar formados por letras, números, espacios y 
puntuación, pero no distinguen entre mayúsculas y minúsculas. Por ejemplo, estos dos 
enlaces: 

	[texto del enlace] [a] 
	[texto del enlace] [A] 

son equivalentes. 

El acceso directo * nombre de enlace implícito * le permite omitir el nombre del 
enlace, en cuyo caso el texto del enlace se usa como nombre. 
Simplemente use un conjunto vacío de corchetes, por ejemplo, para vincular la palabra
"Google" en el sitio web google.com, simplemente puede escribir: 

	[Google] [] 

Y luego definir el enlace: 

	[Google]: http://google.com/ 

Debido a que los nombres de los enlaces pueden contener espacios, este acceso directo incluso funciona para 
varias palabras en el texto del enlace: 

	Visite [Daring Fireball] [] para obtener más información. 

Y luego defina el enlace: 
	
	[Daring Fireball]: http://daringfireball.net/ Las 

definiciones de enlaces se pueden colocar en cualquier lugar de su documento de Markdown. Yo 
tiendo a ponerlos inmediatamente después de cada párrafo en el que están 
utilizados, pero si lo desea, puede poner a todos al final de su 
documento, algo así como notas al pie. 

Aquí hay un ejemplo de enlaces de referencia en acción:

    Recibo 10 veces más tráfico de [Google] [1] que de 
    [Yahoo] [2] o [MSN] [3]. 

      [1]: http://google.com/ "Google" 
      [2]: http://search.yahoo.com/ "Yahoo Search" 
      [3]: http://search.msn.com/ "MSN Search " 

Al utilizar el acceso directo al nombre del enlace implícito, podría escribir: 

    Recibo 10 veces más tráfico de [Google] [] que de 
    [Yahoo] [] o [MSN] []. 

      [google]: http://google.com/ "Google" 
      [yahoo]: http://search.yahoo.com/ "Yahoo Search" 
      [msn]: http://search.msn.com/ "MSN Search "Los 

dos ejemplos anteriores producirán la siguiente salida HTML: 

    <p>
    title = "Google"> Google </a> que de 
    <a href="http://search.yahoo.com/" title="Yahoo Search"> Yahoo </a> 
    o <a href = "http: / /search.msn.com/ "title =" MSN Search "> MSN </a>. </p> 

A modo de comparación, este es el mismo párrafo escrito con 
el estilo de enlace en línea de Markdown: 

    obtengo 10 veces más tráfico de [Google] (http://google.com/ "Google") 
    que de [Yahoo] (http://search.yahoo.com/ "Yahoo Search") o 
    [MSN] (http://search.msn.com/ " Búsqueda de MSN "). 

El punto de referencia de los enlaces de estilo no es que sean más fáciles de 
escribir. El punto es que con enlaces de estilo de referencia, 
Su fuente de documentos es mucho más legible. Compara los ejemplos anteriores: usando
enlaces de estilo de referencia, el párrafo en sí tiene solo 81 caracteres 
; con enlaces de estilo en línea, es de 176 caracteres; y como HTML en bruto, 
tiene 234 caracteres. En el HTML en bruto, hay más marcado que 
texto. 

Con los enlaces de estilo de referencia de Markdown, un documento de origen 
se parece mucho más a la salida final, como se representa en un navegador. Al 
permitirle mover los metadatos relacionados con el marcado fuera del párrafo, 
puede agregar enlaces sin interrumpir el flujo narrativo de su 
prosa.
